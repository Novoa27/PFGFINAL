```python
df_bitcoin = pd.read_csv('Datos/btc_price_hourly.csv')
df_bitcoin.head(5)
```

```python
#datos a redondear: Open, High, Low, Close
redondear = ['Open', 'High', 'Low', 'Close']
for i in redondear:
    print ("Redondeando...:",i)
    df_bitcoin[redondear] = df_bitcoin[redondear].round(2)
#eliminar Dividends, Stock Splits
eliminacion = ['Dividends', 'Stock Splits']
for i in eliminacion:
    print("Eliminando...",i)
    del df_bitcoin[i]
```

```python
#anadir columna hora
#crear lista horas
date_act = "0"
horas = []
hora_act = 0
for i in range (df_bitcoin.shape[0]):
    if df_bitcoin.Date[i] == date_act:
        hora_act += 1
        horas.append(hora_act)
    else:
        date_act = df_bitcoin.Date[i]
        hora_act = 0
        horas.append(hora_act)
#meter lista en df
df_bitcoin['Time'] = horas
```

```python
#crear dataset simple con solo Date, Open
data_for_timeseries = [df_bitcoin['Date'], df_bitcoin['Open'],df_bitcoin['Close']]
headers = ["Date","Open","Close"]
df_for_timeseries = pd.concat(data_for_timeseries, axis=1, keys=headers)
```

```python
df_for_timeseries["Diference"]= np.round((1-(df_for_timeseries["Open"]/df_for_timeseries["Close"]))*100,3)
```

```python
df_for_timeseries.Diference[i]
```

```python
arr_difs = [] #list_difs.append()
for i in range (df_for_timeseries.shape[0]):
    if df_for_timeseries.Diference[i] <= -0.5:
        arr_difs.append("tendencia_bajista")
    if df_for_timeseries.Diference[i] >= +0.5:
        arr_difs.append("tendencia_alcista")
    if df_for_timeseries.Diference[i] <+0.5 and df_for_timeseries.Diference[i] > -0.5:
        arr_difs.append("sin_tendencia")
```

```python
df_for_timeseries["Tendencia"] = arr_difs
```

```python
from keras.layers import Input, Dense, Bidirectional, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Model
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
```

```python
def create_model():
    input_layer = Input(batch_shape=(None, len(x), 1))
    layer = Bidirectional(LSTM(128, return_sequences=True))(input_layer)
    layer = Bidirectional(LSTM(128))(layer)
    out = Dense(1, activation="sigmoid")(layer)
    m = Model(inputs=input_layer, outputs=out)
    m.compile("rmsprop", loss='mean_squared_error')
    return m


def train_model(model, X, y):
    ea = EarlyStopping(monitor='val_loss', patience=2)
    val_loss = model.fit(X, y, epochs=500, batch_size=64, callbacks=[ea], verbose=1, validation_split=.1)
    return val_loss
```

```python
#tendencias
# bitcoin.csv, blockchain.csv, cryptocurrency.csv, ethereum.csv, mining.csv
path="Datos/trends/"
tendencias = ["bitcoin.csv", "blockchain.csv", "cryptocurrency.csv", "Ethereum.csv", "mining.csv","Nvidia.csv"]
j = 1
df_trend = pd.DataFrame()
df_trend['Date'] = pd.read_csv(path+tendencias[0], delimiter = ';').Date
for i in tendencias:
    trend = pd.read_csv(path+i, delimiter = ';')
    df_trend[""+ i] = trend.iloc[:,0]
    j+=1
```

```python
aux = df_trend.replace({'"':''}, regex=True)
aux = aux.replace({'Z':''}, regex=True)
aux = aux.replace({':00.000':''}, regex=True)
aux = aux.replace({'T':' '}, regex=True)
aux = aux.replace({'00:00':''}, regex=True)
```
